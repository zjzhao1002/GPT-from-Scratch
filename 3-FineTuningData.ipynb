```python
from datasets import load_dataset
import json
```

```python
ds = load_dataset("rubenroy/GammaCorpus-Fact-QA-450k", split="train", streaming=True)
```

```python
ds_shuffle = ds.shuffle(seed=42, buffer_size=1000)
ds_finetune = ds_shuffle.take(200)
```

```python
start_question = "<question>"
end_question = "</question>"
start_answer = "<answer>"
end_answer = "</answer>"
```

```python
text_finetune = []
for row in ds_finetune:
    question = row['question']
    answer = row['answer']
    text = f"{start_question}{question}{end_question}{start_answer}{answer}{end_answer}"
    text_finetune.append(text)

print(len(text_finetune))
print(text_finetune[0])
```

```python
with open("fine_tuning.json", 'w', encoding='utf-8') as f:
    json.dump(text_finetune, f, ensure_ascii=False, indent=4)
```
